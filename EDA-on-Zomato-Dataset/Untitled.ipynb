```python
## Therefore.,;; combining both the dataframes.,;;:-:-:_:_:-:-:_:_
final_df=pd.merge(df,df_country,on='Country Code', how='left')
final_df
```

```python
final_df.Country.value_counts()
## Observations:-
## Therefore; from the above; we can conclude that in the US; Zomato is there only in terms of a website that suggests resturants..!!..,,;;
## Whereas.,,;; in India they have a much bigger business with the delivery app as well.,;; therefore this can be concluded from the below output step:-:-:_:_:-:-:_:_..!!..,,;;..!!..,,;;

```

```python
country_names=final_df.Country.value_counts().index
country_names
```

```python
country_values=final_df.Country.value_counts().values
country_values
```

```python
## Creating a Pie Chart:-:-:_:_..!!..,,;; - top 3 counties (sjown below)
plt.pie(country_values[:3],labels=country_names[:3],autopct="%1.2f%%")
plt.show()
```

<u><b>Observations:-</b></u> Zomato maximum records or transactions are from india, followed by US and Uk..!!..,,;;;;;;;;;;;;;;

```python
## getting an observations on the ratings columns:-.,;;
ratings=final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
ratings
```

#### Conclusions/Observations:-.,;;:-:_:_
1. When the rating is 4.5 to 4.9 --> Rtaing:- Excellent.
2. When the rating is 4.0 to 4.4 --> Rtaing:- Very Good.
3.  When the rating is 3.5 to 3.9 --> Rtaing:- Good.
4. When the rating is 2.5 to 3.4 --> Rtaing:- Average.
5. When the rating is 1.8 to 2.4 --> Rtaing:- Poor.
6. When the rating is 0.0 --> Rtaing:- Not Rated.

```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,6)
sns.barplot(x="Aggregate rating", y="Rating Count", data=ratings)
```
